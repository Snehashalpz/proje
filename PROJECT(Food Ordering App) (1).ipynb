```python
# Admin will have the following functionalities: 

 #1. Add new food items. Food Item will have the following details:
        #FoodID //It should be generated automatically by the application.
        #Name
        #Quantity. For eg, 100ml, 250gm, 4pieces etc
        #Price
        # Discount
        #Stock. Amount left in stock in the restaurant.

#2. Edit food items using FoodID.

#3. View the list of all food items.

# 4. Remove a food item from the menu using


# ------------------------------- User ----------------------------------
#The user will have the following functionalities: 

# 1. Register on the application. Following to be entered for registration:
        # Full Name
         #Phone Number
         #Email
         #Address
         #Password

#Log in to the application

#The user will see 3 options:
        # Place New Order
        # Order History
        # Update Profile

# Place New Order: The user can place a new order at the restaurant.
        # Show list of food. The list item should as follows:
            # Tandoori Chicken (4 pieces) [INR 240]
            #Vegan Burger (1 Piece) [INR 320]
            #Truffle Cake (500gm) [INR 900]

#Users should be able to select food by entering an array of numbers. For example, 
#if the user wants to order Vegan Burger and Truffle Cake they should enter [2, 3]

#Once the items are selected user should see the list of all the items selected. 
#The user will also get an option to place an order.
# Order History should show a list of all the previous orders

#Update Profile: the user should be able to update their profile
```

```python
class Admin:
    food_id_counter = 1
    food_items = []

    def add_food_item(self, name, quantity, price, discount, stock):
        food_item = {
            "FoodID": self.food_id_counter,
            "Name": name,
            "Quantity": quantity,
            "Price": price,
            "Discount": discount,
            "Stock": stock
        }
        self.food_id_counter += 1
        self.food_items.append(food_item)
        print("Food item added successfully.")

    def edit_food_item(self, food_id, name, quantity, price, discount, stock):
        for food_item in self.food_items:
            if food_item["FoodID"] == food_id:
                food_item["Name"] = name
                food_item["Quantity"] = quantity
                food_item["Price"] = price
                food_item["Discount"] = discount
                food_item["Stock"] = stock
                print("Food item updated successfully.")
                return
        print("Food item not found.")

    def view_food_items(self):
        for food_item in self.food_items:
            print(food_item)

    def remove_food_item(self, food_id):
        for food_item in self.food_items:
            if food_item["FoodID"] == food_id:
                self.food_items.remove(food_item)
                print("Food item removed successfully.")
                return
        print("Food item not found.")


class User:
    user_id_counter = 1
    users = []

    def register(self, full_name, phone_number, email, address, password):
        user = {
            "UserID": self.user_id_counter,
            "Full Name": full_name,
            "Phone Number": phone_number,
            "Email": email,
            "Address": address,
            "Password": password,
            "Order History": []
        }
        self.user_id_counter += 1
        self.users.append(user)
        print("User registered successfully.")

    def login(self, email, password):
        for user in self.users:
            if user["Email"] == email and user["Password"] == password:
                print("Login successful.")
                return
        print("Invalid credentials.")

    def place_new_order(self):
        print("List of food items:")
        food_items = [
            {"FoodID": 1, "Name": "Tandoori Chicken", "Quantity": "4 pieces", "Price": 240},
            {"FoodID": 2, "Name": "Vegan Burger", "Quantity": "1 Piece", "Price": 320},
            {"FoodID": 3, "Name": "Truffle Cake", "Quantity": "500gm", "Price": 900}
        ]
        for food_item in food_items:
            print(f"{food_item['Name']} ({food_item['Quantity']}) [INR {food_item['Price']}]")

        selected_items = input("Select food items by entering the numbers (comma-separated): ")
        selected_items = list(map(int, selected_items.split(",")))

        order = []
        total_cost = 0
        for food_id in selected_items:
            for food_item in food_items:
                if food_item["FoodID"] == food_id:
                    order.append(food_item)
                    total_cost += food_item["Price"]
                    break

        print("Selected food items:")
        for item in order:
            print(f"{item['Name']} ({item['Quantity']}) [INR {item['Price']}]")
        print("Total Cost:", total_cost)

        place_order = input("Do you want to place the order? (yes/no): ")
        if place_order.lower() == "yes":
            # Implement order placement logic here
            order_details = {
                "OrderID": len(self.users[self.user_id_counter - 2]["Order History"]) + 1,
                "Order Items": order,
                "Total Cost": total_cost
            }
            self.users[self.user_id_counter - 2]["Order History"].append(order_details)
            print("Order placed successfully.")

    def order_history(self):
        user_id = self.user_id_counter - 1
        for user in self.users:
            if user["UserID"] == user_id:
                order_history = user["Order History"]
                if len(order_history) > 0:
                    print("Order History:")
                    for order in order_history:
                        print(order)
                else:
                    print("No order history available.")
                return
        print("User not found.")

    def update_profile(self, new_full_name):
        user_id = self.user_id_counter - 1
        for user in self.users:
            if user["UserID"] == user_id:
                user["Full Name"] = new_full_name
                print("Profile updated successfully.")
                return
        print("User not found.")


# Example usage:

admin = Admin()
admin.add_food_item("Tandoori Chicken", "4 pieces", 240, "5%", 50)
admin.add_food_item("Vegan Burger", "1 Piece", 320, "10%", 30)
admin.add_food_item("Truffle Cake", "500gm", 900, "15%", 20)
admin.view_food_items()
admin.edit_food_item(2, "Vegan Burger Deluxe", "1 Piece", 350, "10%", 30)
admin.view_food_items()
admin.remove_food_item(3)
admin.view_food_items()

user = User()
user.register("John Doe", "1234567890", "johndoe@example.com", "123 Street, City", "password")
user.login("johndoe@example.com", "password")
user.place_new_order()
user.order_history()
user.update_profile("John Smith")
user.order_history()
```
